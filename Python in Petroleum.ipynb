```python
#The permeability of a core plug is measured by air. Only one measurement is made at mean pressure of 2.152psi. The air permeability is 46.6md
#Estimate the absolute permeability of the core sample. Compare with 22.66md
```

```python
k = input('\n Enter the first guess of absolute permeability kl:')
k = float(k)

pm = input('\n Enter mean pressure pm:')
pm = float(pm)

kg = input('\n Enter k(air) kg:')
kg = float(kg)

while abs(0.9*(k**0.64) + pm*k - pm*kg) > 0.01:
    k = k -(0.9*(k**0.64) + pm*k - pm*kg)/(4.412*(k**(0.36)) * pm)
    
print(f'The final value of Perm K is:  {k}')
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline

#We get the IPR set
q_ipr = [0, 1999,3456, 3982, 4512, 4963, 5275, 5468,5519]
pwf_ipr = [4000,3500,3000,2500,2000,1500,1000,500,14.7]

ipr = {'Pwf': [4000,3500,3000,2500,2000,1500,1000,500,14.7],
      'Q': [0, 1999,3456, 3982, 4512, 4963, 5275, 5468,5519]}
ipr_df = pd.DataFrame({'Pwf': [4000,3500,3000,2500,2000,1500,1000,500,14.7],
      'Q': [0, 1999,3456, 3982, 4512, 4963, 5275, 5468,5519]})

ipr_df
```

```python
#Let us look at TPR Curve

q_tpr = [1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000]
p_190 = [1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2609]
p_2375 = [1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]
p_2875 = [1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]

tpr = {'q': q_tpr, 'Pwf(1.90tbg)': p_190, 'Pwf(2.375tbg)': p_2375, 'Pwf(2.875tbg)': p_2875}

plt.figure(figsize=(8,4))

plt.style.use('fivethirtyeight')
plt.plot(ipr['Q'], ipr['Pwf'], label ='IPR',linewidth = 1.5)

plt.plot(tpr['q'], tpr['Pwf(1.90tbg)'],label = 'TPR for 1.9" tubing', linewidth = 5 )
plt.plot(tpr['q'], tpr['Pwf(2.375tbg)'],label = 'TPR for 2.3" tubing', linewidth = 5 )
plt.plot(tpr['q'], tpr['Pwf(2.875tbg)'],label = 'TPR for 2.8" tubing', linewidth = 5 )

plt.xlabel('FlowRate(Mscf/D)')
plt.ylabel('Flowing BHP(Psia)')
plt.title('Nodal Analysis for a Gas Well')
plt.grid()
plt.legend(loc= 'best')
```
